```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading data
data=sns.load_dataset("tips")
#top 5 data points
data.head()
```

```python
sns.heatmap(data.corr())
plt.show()
```

```python
data.corr()
```

```python
sns.heatmap(data.corr(),annot=True,cmap='PuBuGn')
plt.show()
```

```python
sns.heatmap(data.corr(),annot=True,cmap='PiYG_r')
plt.show()
```

```python
sns.boxenplot(x='smoker',y='total_bill',data=data)
plt.show()
```

```python
#it presents five number summary
sns.boxplot(x='smoker',y='total_bill',data=data)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=data,palette='rainbow')
```

```python
sns.boxplot(x='total_bill',y='day',data=data,hue='smoker')
```

```python
sns.violinplot(x='total_bill',y='day',data=data)
```

```python
#it is a combination of boxplot and kernel density distribution/probability distribution
sns.violinplot(x='total_bill',y='day',data=data)
```

```python
sns.violinplot(x='total_bill',y='day',data=data,hue='smoker',palette='husl')
```

```python
sns.pairplot(data,diag_kind='kde')
```

```python
sns.pairplot(data,diag_kind='hist')
```
